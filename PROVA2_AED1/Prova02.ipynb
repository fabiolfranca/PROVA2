```python
import numpy as np

consumo_diario = [ 50, 45, 55, 60, 48, 52, 47, 49, 53, 51, 50, 46, 54, 
                  61, 49, 53, 48, 50, 52, 47, 56, 50, 55, 60, 48, 52,
                  47, 49, 53, 51]
```

```python
def consumo(consumo_diario):
    media = np.mean(consumo_diario)
    return media

```

```python
print(f'o consummo diário médio foi: {consumo(consumo_diario)}')
```
